```python
from duckduckgo_search import ddg_images
from fastcore.all import *

def search_images(term, max_images=30):
    print(f"Searching for '{term}'")
    return L(ddg_images(term, max_results=max_images)).itemgot('image')
```

```python
#NB: `search_images` depends on duckduckgo.com, which doesn't always return correct responses.
#    If you get a JSON error, just try running it again (it may take a couple of tries).
urls = search_images('bird photos', max_images=1)
urls[0]
```

```python
download_url(search_images('forest photos', max_images=1)[0], 'forest.jpg', show_progress=False)
Image.open('forest.jpg').to_thumb(256,256)#to_thumb is from fast.ai
```

```python
searches = 'forest','bird'
path = Path('bird_or_not')
from time import sleep

for o in searches:
    dest = (path/o)
    dest.mkdir(exist_ok=True, parents=True)
    download_images(dest, urls=search_images(f'{o} photo'))
    sleep(1)  # Pause between searches to avoid over-loading server
    download_images(dest, urls=search_images(f'{o} sun photo'))
    sleep(1)
    download_images(dest, urls=search_images(f'{o} shade photo'))
    sleep(1)
    resize_images(path/o, max_size=400, dest=path/o)#we do not need big images so we're resizing
```

```python
failed = verify_images(get_image_files(path))
failed.map(Path.unlink)
len(failed)
```

```python
dls = DataBlock(#how to use different kind of datasets , we can basically split into these 5 things
    blocks=(ImageBlock, CategoryBlock), #what kind of input is givern(image block) and what is output category
    get_items=get_image_files, #returns a list of all of the image files
    splitter=RandomSplitter(valid_pct=0.2, seed=42),#for splitting into test and train (randomly set aside 20% of data)
    get_y=parent_label,#how do we know the correct label of a photo simply retruns parent folder 
    item_tfms=[Resize(192, method='squish')]#all of inputs need to be of same size (192*192) by squishing
).dataloaders(path, bs=32)#dataloaders is used to iterate through a batch of data at a time
dls.show_batch(max_n=6)#show batch 
```

```python
#model learner is used combine the neural network with dataset so two arguments
#dls and resnet18 #you can use any of the pytorch image models (timm)
learn = vision_learner(dls, resnet18, metrics=error_rate)
learn.fine_tune(3)#takes pretrained weights adjusts them and teaches the model the differences 
```
